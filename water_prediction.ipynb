```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['class'] = le.fit_transform(df1['class'])
#encoded_classes = le.fit_transform(class)
print(df1)
```

```python
from scipy import stats

# Calculate the Z score for each data point in the dataset
z_scores = stats.zscore(df1)

# Print the Z score for each data point
print(z_scores)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df1, x='PH', y='conductivity', hue='class')
plt.title('Scatter Plot: pH vs Conductivity')
plt.show()
```

```python
subset = ['PH', 'conductivity', 'BOD', 'class']
sns.pairplot(data=df1[subset], hue='class')
plt.title('Pairplot: PH, Conductivity, BOD')
plt.show()
```

```python
y=df1.iloc[:,22]
print(y)

X=z_scores.iloc[:,0:22]
print(X)
```

#Machine Learning Algorithm (Logistic Regression, Decision Tree, Random Forest Tree )

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=25)
from numpy import mean
from numpy import std
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import classification_report

model = LogisticRegression(multi_class='multinomial', solver='newton-cg').fit(X_train, y_train)
preds = model.predict(X_test)
# define the model evaluation procedure
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate the model and collect the scores
n_scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# report the model performance
print('Mean Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))

params = model.get_params()
#print(params)
print('Logistic regression')
class_report=classification_report(y_test, preds)
print(class_report)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=10)

clf = DecisionTreeClassifier(max_depth=6, min_samples_split=2) #RS=35 FOR ACCURACY DATA SET=80
#RS=10-->85
clf = clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print("decision tree")
print(classification_report(y_test,y_pred))



plt.figure(figsize=(20,10))
plot_tree(clf, filled=True)
plt.show()
```

```python

from sklearn.ensemble import RandomForestClassifier
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
#Randfomforest Classifier
clf=RandomForestClassifier(n_estimators=250)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print("randomforest")
print(classification_report(y_test,y_pred))
```
